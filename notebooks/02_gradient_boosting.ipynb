# Gradient Boosting

Gradient Boosting - це метод в машинному навчанні, який об'єднує багато слабких моделей у одну сильну (Кілька трошки тупих моделей разом стають powerful розумними)

_Зазвичай використовується для прогнозів і класифікацій_

### Проблема
- Чому проста модель (одне дерево) - це часто погано?
  1) Дерева можуть помилятися, вгадувати не точно
  2) Одне дерево можу бути недонавченим/перенавченим
  3) Хочеться отримати точнішу модель, без створення "ДУЖН ВЕЛИКОГО" дерева

### Як працює Gradient Boosting
- Gradient boosting - це Boosting, який використовує градієнт (тіп напрямок у якому треба змінити передбачення, щоб найшвидше зменшити помилку)
- Кожна нова модель вчиться не просто на помилках, а на тому, наскільки і в який бік треба змінити прогноз, щоб зменшити втрати
1) Починаємо з простої модельки
2) Визначаємо помилки (різниця між прогнозом і реальними данними)
3) Наступна модель навчається на цих помилках, щоб їх виправити
4) Процес повторються послідовно крок за кроком
5) всі моделі об'єднуються у сумарну прогностичну модель

карта помилок - це тіп список/масив, де для кожного прикладу сказано наскільки сильно ми помилились. Наступна модель дивиться на цю карту і вчиться на помилках, виправляючи їх

### Переваги Gradient Boosting

1) Висока точність прогнозування
2) Може обробляти різні типи даних
3) Добре працює з несбалансованими даними

### Недоліки Gradient Boosting
1) Може переобучитися, якщо дерев багато
2) Вимагає більше часу на навчання

### Метафора для розуміння
- Gradient Boosting - це як команда мопсів, але тепер вони не просто бачать свої помилки, ВОНИ ще й РОЗУМІЮТЬ, наскільки сильно кожна помилка псує результат
  1) Перший мопс приносить м'яч, як може, але помиляється
  2) Другий мосп не просто бачить, які м'ячі пропущені, він бачить наскільки важливий кожен пропущений м'яч
  3) Третій мопс фокусується на тих м'ячих, які створюють найбільші проблеми
  4) в результаті команда мопсіків вчиться СУПЕРТОЧНО приносити м'ячі, бо кожна наступна модель вчиться тіп на градієнтах помилок

### Як це працює (покроково)
1) Модель №1 робить перший прогноз.
2) Дивимось де воно помилилося (карта помилок)
3) Друга модель вчиться саме на цих помилках
4) Додаємо роботу другої моделі до першої
5) повторюємоо

### Чим Gradient Boosting відрізняється від Boosting
- Замість цього модель вчиться на градієнтах функції втрат (крутіше, гнучкіше, точніше)
- Працює з будь-якими функціями втрат.


крч, Gradient Boosting - це спосіб перетворити багато маленькх моделей у дуже сильну модель.
Головна ідея:
кожна нова модель вчитсься на тому, як зменшити помилку на попередніх(на градієнтах втрат)


### Gradient Boosting Classifier - що це таке?
- Gradient Boosting - це ідея/алгоритм (метод тренерування команди мопсав)
- GradientBoostingClassifier - це конкретна модель, яка застосовує GradientBoosting для задач класифікації (команда мопсав, навчена вирішувати питання "малий чи велкий м'яч" або "червоний чи синій")
